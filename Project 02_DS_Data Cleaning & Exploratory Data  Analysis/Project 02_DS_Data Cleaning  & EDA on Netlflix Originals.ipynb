```python
# Importing libraries that are needed 
# Importing pandas library
import pandas as pd
# Specify the encoding
df = pd.read_csv("C:\\Users\\Shiva Goud\\Desktop\\NetflixOriginals.csv", encoding='ISO-8859-1')
display(df)
```

```python
# Data Set Details - Info
df.info()
```

```python
# Handling Missing Values
# displaying null values
df.isnull().sum()
```

```python
# Duplicate Values
df.duplicated().sum()
```

```python
# Displaying Shape
df.shape
```

```python
df.columns
```

```python
# Data Types
df.select_dtypes(include='object').describe()  
```

```python
df['Premiere'] = pd.to_datetime(df['Premiere'].str.replace('.', ',', regex=False), format='%B %d, %Y').dt.strftime('%d-%m-%Y')
```

```python
df['Premiere'] = pd.to_datetime(df['Premiere'], format='%d-%m-%Y')
```

```python
df['Premiere'].info()
```

```python
df
```

```python
df[['Primary_Genre', 'Genre2', 'Genre3', 'Genre4']] = df['Genre'].str.split('/', expand=True)
df
```

```python
df[['Primary_Language', 'Language2', 'Language3']] = df['Language'].str.split('/', expand=True)
df
```

```python
df.drop(['Genre2', 'Genre3', 'Genre4', 'Language2', 'Language3'], axis=1, inplace=True)
df
```

```python
# importing seaborn library
import seaborn as sns
# Heat Map for null values
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap="Blues")
```

```python
# Average IMDB score for each genre
df.groupby('Primary_Genre')['IMDB Score'].mean().sort_values(ascending=False).round(decimals=1).reset_index()
```

```python
# Importing Matplotlib library
import matplotlib.pyplot as plt

# Data Visualizing IMDB Score VS Primary Genre
imdb = df.groupby('Primary_Genre')['IMDB Score'].mean().sort_values(ascending=False).round(decimals=1).head(10)
plt.figure(figsize=(10,5),dpi=90, frameon=True)
# Assigning x label & y label
plt.ylabel('Primary_Genre')
plt.xlabel('IMDB Score')
plt.title('Top 10 Genre by IMDB Score', size=14)

imdb.plot.barh(color='green')
plt.show()
```

```python
# Genre has the highest average runtime
df.groupby('Primary_Genre')['Runtime'].mean().sort_values(ascending=False).round().reset_index()
```

```python
# Data Visualizing Run time VS Primary Genre
avg_run = df.groupby('Primary_Genre')['Runtime'].mean().sort_values(ascending=False).round().head(10)
plt.figure(figsize=(14, 6))
sns.barplot(x=avg_run.values, y=avg_run.index, orient='h', palette='viridis')

# Assigning Title ,x-label & y-label
plt.title('Highest Average Time By Genre', size=13)
plt.xlabel('Runtime (minutes)')
plt.ylabel('Primary Genre')
plt.show()
```

```python
# longer movies rated higher on IMDB
# corelation between movies runtime & rating
plt.figure(figsize=(14, 6), dpi=90, frameon=True)
sns.scatterplot(data= df, x='Runtime', y='IMDB Score', hue='IMDB Score', palette='autumn_r')
plt.title('correlation between movies runtime and its rating', size=13)
plt.show()
```

```python
# the distribution of movies across different languages
df['Primary_Language'].value_counts(ascending=False)
```

```python
lang = df['Primary_Language'].value_counts(ascending=False)
plt.figure(figsize=(15, 7))
sns.barplot(x=lang.values, y=lang.index, palette='prism')

# Assigning Title , x-label & y-label
# Plotting a bar graph counts vs languages
plt.title('Distribution of movies accross different Languages', fontsize=15)
plt.xlabel('Counts')
plt.ylabel('Languages')
plt.show()
```

```python
# The trend in the number of releases over the years
#  Grouping the data by year and count the number of movies that were released in every year

df['Premiere'] = pd.to_datetime(df['Premiere'])
df['Year'] = df['Premiere'].dt.year
grouped = df.groupby('Year')['Title'].count()
print(grouped)
```

```python
# Plotting a graph of grouped data
grouped.plot(figsize=(15,5))
plt.plot(grouped.index, grouped.values, color='blue', marker='o')   

# Assigning x-label & y-label
plt.ylabel('No. of Movie Releases')
plt.title('Distribution of Movie Releases Over the Years', size=14)

for i, value in enumerate(grouped.values):
    plt.annotate(str(value), (grouped.index[i], value), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
# Year had the highest average IMDB score
df.groupby('Year')['IMDB Score'].mean().sort_values(ascending=False).round(decimals=1)
```

```python
# plotting the  bar graph
yr = df.groupby('Year')['IMDB Score'].mean().round(decimals=1)
plt.figure(figsize=(13, 5), dpi=90)
sns.barplot(x=yr.index, y=yr.values, palette='inferno', width=0.5)
plt.title('Average IMDB Score by Year', size=13)
plt.ylabel('IMDB Score')

plt.show()
```

```python
# movies in certain languages rated higher on average
df.groupby('Primary_Language')['IMDB Score'].mean().sort_values(ascending=False).round(decimals=1)
```

```python
pg_lang = df.groupby('Primary_Language')['IMDB Score'].mean().sort_values(ascending=False).round(decimals=1)

# Plotting Graph of high rated movies by languages 
plt.figure(figsize=(14, 7),dpi=80, frameon=True)
sns.barplot(x=pg_lang.values, y=pg_lang.index, palette='cool')

# Assigning title x-label,y-label & Title
plt.title('High Rated Movies by Languages', size=14)
plt.ylabel('Languages')
plt.xlabel('IMDB Score')

plt.show()
```

```python
# The distribution of IMDB scores
fig, ax = plt.subplots(figsize=(10, 4))
sns.histplot(data=df, x='IMDB Score', bins=5, ax=ax, color='Orange')
plt.title('Histogram of IMDB Scores', size=14)
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# The most common genre on Netflix
df['Primary_Genre'].value_counts().sort_values(ascending=False)
```

```python

# Plotting graph 
genre = df['Primary_Genre'].value_counts().sort_values(ascending=False).head()
plt.pie(genre.values, labels= genre.index, autopct='%1.1f%%', startangle=90, explode=[0, 0.1, 0.1, 0.1, 0.1])

# Assigning x-label , y-label & Title
plt.title('Top 5 Most Common Genre', size=13)
plt.axis('equal')
plt.show()
```

```python
#  the average runtime of movies released in different months
df['Premiere'] = pd.to_datetime(df['Premiere'])
df['Month'] = df['Premiere'].dt.month
groupedM = df.groupby('Month')['Runtime'].mean().round()
print(groupedM)
```

```python
plt.figure(figsize=(16, 6), dpi=90, frameon=True)

sns.lineplot(x=groupedM.index, y=groupedM.values, marker='o', color='orange')

# Asssigning x-label ,y-label & Title
plt.title('Average Runtime of Movies by Month', size=15)
plt.ylabel('Runtime (minutes)')

for i, value in enumerate(groupedM.values):
    plt.annotate(str(value), (groupedM.index[i], value), textcoords="offset points", xytext=(0,12), ha='center')


plt.show()
```
